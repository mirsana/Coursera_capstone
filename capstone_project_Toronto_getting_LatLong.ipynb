# Segmenting and Clustering Neighborhoods in Toronto

###  import website scraping libraries and packages in Python. 

```python
import numpy as np
import pandas as pd

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### scraping data from wikipedia

```python
def getHTMLContent(link):
    html = urlopen(link)
    soup = BeautifulSoup(html, 'html.parser')
    return soup
```

```python
content = getHTMLContent('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
tables = content.find_all('table')
for table in tables:
    print(table.prettify())
```

```python
A=[]
B=[]
C=[]
table = content.find('table', {'class': 'wikitable sortable'})
rows = table.find_all('tr')

for row in rows:
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
        
```

###  you can simply use pandas to read the table into a pandas dataframe.

```python

df=pd.DataFrame(A,columns=['Postal code'])
df['Borough']=B
df['Neighborhood']=C
```

```python
df.head()
```

# Data Wrangling

```python
df.dtypes
```

###  Ignore cells with a borough that is Not assigned.

```python
 df['Borough'].isnull().sum()
```

```python
df['Borough'].replace('Not assigned\n', np.nan, inplace=True)
```

```python
df['Borough'].isnull().sum()
```

```python
df.dropna(subset=["Borough"], axis=0, inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

### if more than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma as shown in row 11 in the above table.

```python
temp_df=df.groupby('Postal code')['Neighborhood'].apply(lambda x:', '.join(x))
temp_df=temp_df.reset_index(drop=False)
temp_df.rename(columns={'Neighborhood':'Neighborhood_joined'},inplace=True)
```

```python
df_merge = pd.merge(df, temp_df, on='Postal code')
df_merge.drop(['Neighborhood'],axis=1,inplace=True)
df_merge.drop_duplicates(inplace=True)
df_merge.rename(columns={'Neighborhood_joined':'Neighborhood'},inplace=True)
df_merge.head()
```

```python
df=df_merge
```

```python
df.head()
```

```python
df['Neighborhood'] = df['Neighborhood'].str.strip()
df.head()
```

### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.


```python
df.loc[df['Neighborhood']== 'Not assigned\n',['Neighborhood']] = df.loc[df['Neighborhood']== 'Not assigned\n',['Borough']]
```
